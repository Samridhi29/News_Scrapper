```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'}
webpage=requests.get("https://www.hindustantimes.com/entertainment/bollywood/kangana-ranaut-defends-pm-modi-amid-trolling-101694580352675.html",headers=headers).text 
```

```python
soup=BeautifulSoup(webpage, 'lxml')
```

```python
paragraphs=soup.find_all('p')
sub_head=soup.find_all('h2')
header=soup.find_all('h1')
```

```python
paragraph_texts = [p.get_text() for p in paragraphs]
sub_headings=[a.get_text() for a in sub_head]
heading=[h.get_text() for h in header]
```
